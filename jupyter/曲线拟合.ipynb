```python
x=np.arange(1.1,4.1,0.1)
y=np.array([8.0725, 5.2761, 3.8791, 3.0955, 2.6103, 2.2853, 2.0541, 1.8822, 1.7497, 1.6449, 
            1.5602, 1.4905, 1.4324, 1.3833, 1.3414, 1.3054, 1.2742, 1.2470, 1.2231, 1.2020,
            1.1833, 1.1667, 1.1519, 1.1386, 1.1267, 1.1159, 1.1062, 1.0975, 1.0895, 1.0823 ])
def f(x,a,b,c):
    return np.power((c/1.0/(a*x+b)),0.5)

fita,fitb=optimize.curve_fit(f,x,y)
print(fita)

plt.plot(x,y)
plt.plot(x,f(x,fita[0],fita[1],fita[2]))
# plt.plot(x,f(x,0.5,-0.55,1.6))
plt.show()
```
